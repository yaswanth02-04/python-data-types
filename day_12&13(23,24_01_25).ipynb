```python
# Calculate q1 and q3 (first and third quartiles)
q1 = df['fare'].quantile(0.25)  # Replace 'fare' with the relevant column if needed
q3 = df['fare'].quantile(0.75)  # Replace 'fare' with the relevant column if needed

# Now you can use q1 and q3 in your calculations
lower_bound = q1 - 1.5 * (q3 - q1)
upper_bound = q3 + 1.5 * (q3 - q1)  # Corrected the formula here: 1.5 instead of 1-5

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```
